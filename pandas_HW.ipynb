## Iris dataset

### For who is unaware of bothanic language (like me), here's an example of what petal and sepal are:



```python
# This dataset has been chosen due to its availability from all platforms without any need of framework or .csv file

import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.tail()
```

```python
iris.species = iris.species.astype('category')
```

```python
iris.info()
```

```python
iris.describe()
```

### Bar plot

```python
fig, ax = plt.subplots()

ax.bar(iris.sepal_length, iris.sepal_width, label='Sepal', alpha=0.5)
ax.bar(iris.petal_length, iris.petal_width, label='Petal', alpha=0.5)
ax.legend()

plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Sepal & Petal (width by length)')
```

#### We visually observe a correlation between length and width of petal. While sepal values are greater, there is less significant correlation between its length & width compared to petal. 

```python
plt.style.available
```

### Scatter plot

#### Using scatterplot, we'll visualize the relation between lenth & width within sepal and petal of the species.

```python
plt.style.use('seaborn-pastel')

fig, ax = plt.subplots()

ax.scatter(iris['sepal_length'], iris['sepal_width'], label='Sepal', color='green', linewidths=0.1)
ax.scatter(iris['petal_length'], iris['petal_width'], label='Petal', color='blue', linewidths=0.1)

ax.legend()

ax.set_title('Relation between length & width')
ax.set_xlabel('length')
ax.set_ylabel('width')

plt.tight_layout()

plt.show()
```

### Multiple boxplots

```python
fig, axes = plt.subplots(2,2, sharex=True, figsize=(10,6))

sns.boxplot(x='species', y='sepal_length', data=iris, ax=axes[0][0])
axes[0][0].set(xlabel=None)
sns.boxplot(x='species', y='sepal_width', data=iris, ax=axes[1][0])
sns.boxplot(x='species', y='petal_length', data=iris, ax=axes[0][1])
axes[0][1].set(xlabel=None)
sns.boxplot(x='species', y='petal_width', data=iris, ax=axes[1][1])

fig.suptitle('Some deep visualisation over species with boxplot')
plt.tight_layout()
plt.show()
```

## Groupby

```python
iris.groupby('species').mean()
```

```python
species_group = iris.groupby(["species"])
```

```python
species_group.get_group('setosa').head()
```

```python
species_group.get_group('virginica').head()
```

```python
df1=species_group["sepal_length"].agg(["median", "mean", "std"])
df2=species_group["petal_length"].agg(["median", "mean", "std"])
df3=species_group["sepal_width"].agg(["median", "mean", "std"])
df4=species_group["petal_width"].agg(["median", "mean", "std"])
```

### Sepal & Petal Area

```python
iris['sepal_area']=iris.sepal_length * iris.sepal_width
```

```python
iris['petal_area']=iris.petal_length * iris.petal_width
```

```python
plt.subplots()
plt.plot(np.arange(150), iris.sepal_area.sort_values(), label='Sepal Area')
plt.fill_between(np.arange(150), iris.sepal_area.sort_values(), alpha=0.5)
plt.plot(np.arange(150), iris.petal_area.sort_values(), label='Petal Area')
plt.fill_between(np.arange(150), iris.petal_area.sort_values(), alpha=0.5)

plt.xlabel('No. of observation (cumulative)')
plt.ylabel('cm^2')
plt.legend()
```
